```python
# Import libraries and create dataframe

import pandas as pd
import numpy as np

np.random.seed(42)  
data = pd.DataFrame({
    "name": ["Aragorn", "Boromir", "Frodo", "Galadriel", "Legolas", "Gimli"],
    "age": np.random.randint(18, 900, size=6)  # numeric column
})

data
```

```python
# Create a new column that adds 10 to every age
data['age_plus_10'] = data['age'].apply(lambda x: x + 10)
data
```

```python
# Create a flag column
hobbits = {'Frodo'}
data['is_hobbit'] = data['name'].apply(lambda x: x in hobbits)
data
```

```python
# Save
data.to_csv("20th_century_testdata.csv", index=False)
```
